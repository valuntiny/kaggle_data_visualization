# ploting package

## line chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# read data
fifa_data = pd.read_csv("fifa.csv", index_col='Date', parse_dates=True)
# the parse_dates para trying to parse the date format
fifa_data.head(5)
```

```python
# plot it, first set the figure size
plt.figure(figsize=(16,6))
# get a title
plt.title("random title")
# line chart
sns.lineplot(data=fifa_data)
```

```python
museum_data = pd.read_csv("museum_visitors.csv", index_col='Date', parse_dates=True)
museum_data.head(5)
```

```python
plt.figure(figsize=(14,6))
plt.title("monthly visitors for each museum")
sns.lineplot(data=museum_data)
```

```python
# plot use subdata
plt.figure(figsize=(12, 6))
plt.title("monthly visitors for Avila Adobe")
sns.lineplot(data=museum_data['Avila Adobe'], label='Avila Adobe')
plt.xlabel('Date')  # label x axis
```

## bar chart and heatmap

```python
ign_data = pd.read_csv("ign_scores.csv", index_col="Platform")
ign_data.head(5)
```

```python
# since there are too many platform, let's flip it
plt.figure(figsize=(8,7))
plt.title('barplot')
sns.barplot(data=ign_data, y=ign_data.index, x=ign_data['Racing'])
plt.xlabel('score')
```

```python
plt.figure(figsize=(8,7))
plt.title('heatmap')
sns.heatmap(data=ign_data, annot=True)
plt.xlabel('genre')
```

## scatter plot

```python
candy_data = pd.read_csv('candy.csv', index_col='id')
candy_data.head(5)
```

```python
plt.figure(figsize=(8, 6))
plt.title('scatterplot')
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

```python
# scatterplot with regression line
plt.figure(figsize=(8, 6))
plt.title('scatterplot with regression line')
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

```python
# another way to make it work
sns.lmplot(x='sugarpercent', y='winpercent', data=candy_data)
```

```python
# use hue
plt.figure(figsize=(8, 6))
plt.title('scatterplot')
sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
```

```python
# use hue and regression line
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data=candy_data)
```

```python
sns.swarmplot(x='chocolate', y='winpercent', data=candy_data)
```
